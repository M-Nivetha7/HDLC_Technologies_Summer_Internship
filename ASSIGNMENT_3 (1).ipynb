```python
import numpy as np

A1 = np.array([11, 33, 22, 44, 66, 55])
A2 = np.array([10, 30, 20, 40, 60, 50])

# 1. Extract elements from index 2 to 5
extract = A1[2:6]
print("1. Extracted:", extract)

# 2. Reshape A1 into (3,2)
reshape = A1.reshape(3, 2)
print("2. Reshaped A1:\n", reshape)

# 3. Join A1 and A2
join = np.concatenate((A1, A2))
print("3. Joined array:", join)

# 4. Split A1 into 3 arrays
split = np.array_split(A1, 3)
print("4. Split arrays:", split)

# 5. Search for number 44 in A1
index_44 = np.where(A1 == 44)
print("5. Index of 44:", index_44[0])

# 6. Print all even numbers of A1 using search option
even_nums = A1[np.where(A1 % 2 == 0)]
print("6. Even numbers:", even_nums)

# 7. Sort A2
sorted_A2 = np.sort(A2)
print("7. Sorted A2:", sorted_A2)

# 8. Filter odd numbers in A1 using filter option
odd_nums = A1[A1 % 2 != 0]
print("8. Odd numbers:", odd_nums)
```
