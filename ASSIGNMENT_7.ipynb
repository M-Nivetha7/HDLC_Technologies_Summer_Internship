```
import matplotlib.pyplot as plt
import numpy as np

# -----------------------------
# Sample Data
# -----------------------------
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun']
sales_product_a = [150, 200, 250, 300, 280, 330]
sales_product_b = [180, 220, 260, 310, 300, 360]

student_scores = [65, 70, 75, 80, 85, 90, 95, 100, 60, 55, 50, 45]

categories = ['Electronics', 'Clothing', 'Grocery', 'Toys']
category_values = [120, 150, 90, 60]

marks = {
    'Math': [78, 85, 92, 88, 76],
    'Science': [72, 89, 94, 90, 80],
    'English': [65, 70, 75, 80, 85]
}

# =====================================================
# 1. LINE PLOT
# =====================================================
plt.figure(figsize=(8, 5))
plt.plot(months, sales_product_a, marker='o', label='Product A')
plt.plot(months, sales_product_b, marker='s', label='Product B')

plt.xlabel("Months")
plt.ylabel("Sales")
plt.title("Monthly Sales of Product A and Product B")
plt.legend()
plt.grid(True)
plt.show()

# =====================================================
# 2. GROUPED BAR CHART
# =====================================================
x = np.arange(len(months))
width = 0.35

plt.figure(figsize=(8, 5))
plt.bar(x - width / 2, sales_product_a, width, label='Product A')
plt.bar(x + width / 2, sales_product_b, width, label='Product B')

plt.xticks(x, months)
plt.xlabel("Months")
plt.ylabel("Sales")
plt.title("Monthly Sales Comparison")
plt.legend()
plt.show()

# =====================================================
# 3. HORIZONTAL BAR CHART
# =====================================================
plt.figure(figsize=(8, 5))
bars = plt.barh(categories, category_values)

# Data labels
for bar in bars:
    plt.text(
        bar.get_width() + 2,
        bar.get_y() + bar.get_height() / 2,
        str(int(bar.get_width())),
        va='center'
    )

plt.xlabel("Sales")
plt.ylabel("Categories")
plt.title("Category Sales")
plt.show()

# =====================================================
# 4. SCATTER PLOT
# =====================================================
plt.figure(figsize=(8, 5))

plt.scatter(
    months,
    sales_product_a,
    marker='o',
    label='Product A'
)

plt.scatter(
    months,
    sales_product_b,
    marker='^',
    label='Product B'
)

plt.xlabel("Months")
plt.ylabel("Sales")
plt.title("Product A vs Product B Sales")
plt.legend()
plt.show()

# =====================================================
# 5. HISTOGRAM
# =====================================================
plt.figure(figsize=(8, 5))
plt.hist(student_scores, bins=6)

plt.xlabel("Scores")
plt.ylabel("Frequency")
plt.title("Student Score Distribution")
plt.show()

# =====================================================
# 6. PIE CHART
# =====================================================
largest_index = category_values.index(max(category_values))

explode = [0] * len(categories)
explode[largest_index] = 0.1

plt.figure(figsize=(7, 7))
plt.pie(
    category_values,
    labels=categories,
    autopct='%1.1f%%',
    startangle=90,
    explode=explode
)

plt.title("Category Sales Proportion")
plt.show()

print("Largest Category:")
print(categories[largest_index], "-", category_values[largest_index])

# =====================================================
# 7. SUBPLOTS (2x2)
# =====================================================
fig, axs = plt.subplots(2, 2, figsize=(10, 8))

# Line Plot
axs[0, 0].plot(months, sales_product_a, marker='o')
axs[0, 0].set_title("Product A Sales")

# Bar Chart
axs[0, 1].bar(months, sales_product_b)
axs[0, 1].set_title("Product B Sales")

# Scatter Plot
axs[1, 0].scatter(months, sales_product_a, marker='o')
axs[1, 0].set_title("Product A Scatter")

# Pie Chart
axs[1, 1].pie(
    category_values,
    labels=categories,
    autopct='%1.1f%%',
    startangle=90
)
axs[1, 1].set_title("Category Sales")

plt.tight_layout()
plt.show()

```
