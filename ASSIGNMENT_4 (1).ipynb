```
import numpy as np

def process_array(arr):
    print("Original Array:\n", arr)

    # Row-wise maximum replacement
    row_max = np.max(arr, axis=1, keepdims=True)
    row_replaced = np.repeat(row_max, arr.shape[1], axis=1)
    print("\nRow-wise max replaced:\n", row_replaced)

    # Column-wise minimum replacement
    col_min = np.min(arr, axis=0)
    col_replaced = np.repeat(col_min[np.newaxis, :], arr.shape[0], axis=0)
    print("\nColumn-wise min replaced:\n", col_replaced)


# Example usage
arr = np.random.randint(1, 101, (6, 6))
process_array(arr)
```

```
import numpy as np

def process_image(img):
    print("Original Image Shape:", img.shape)

    # Normalize to [0, 1]
    normalized = img / 255.0
    print("\nNormalized Image (0 to 1):\n", normalized)

    # Multiply by 5
    scaled = img * 5
    print("\nImage multiplied by 5:\n", scaled)


# Example usage
img = np.random.randint(0, 256, (512, 512))
process_image(img)
```
