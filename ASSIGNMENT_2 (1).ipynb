```python
# Complete Updated Python Code for Population Dataset Analysis


# ============================================
# POPULATION DATASET ANALYSIS PROJECT
# ============================================

# Import Required Libraries
import pandas as pd
import numpy as np

# ============================================
# LOAD DATASET
# ============================================

# Load CSV File
file_path = "population.csv"
df = pd.read_csv(file_path)

# ============================================
# DISPLAY COMPLETE DATASET
# ============================================

print("\n========== COMPLETE DATASET ==========")
print(df)

# ============================================
# DISPLAY FIRST ROWS
# ============================================

print("\n========== FIRST 5 ROWS ==========")
print(df.head())

print("\n========== FIRST 10 ROWS ==========")
print(df.head(10))

# ============================================
# DISPLAY LAST ROWS
# ============================================

print("\n========== LAST 5 ROWS ==========")
print(df.tail())

print("\n========== LAST 8 ROWS ==========")
print(df.tail(8))

# ============================================
# DATASET SHAPE
# ============================================

print("\n========== NUMBER OF ROWS ==========")
print(df.shape[0])

print("\n========== NUMBER OF COLUMNS ==========")
print(df.shape[1])

# ============================================
# COUNTRY WITH MAXIMUM POPULATION IN 1970
# ============================================

max_population_country = df.loc[df['1970'].idxmax(), 'Country Name']
max_population_value = df['1970'].max()

print("\n========== COUNTRY WITH MAXIMUM POPULATION IN 1970 ==========")
print("Country Name :", max_population_country)
print("Population   :", max_population_value)

# ============================================
# AVERAGE POPULATION IN YEAR 2000
# ============================================

average_population_2000 = df['2000'].mean()

print("\n========== AVERAGE POPULATION IN YEAR 2000 ==========")
print(average_population_2000)

# ============================================
# DATA TYPES OF EACH COLUMN
# ============================================

print("\n========== DATATYPE OF EACH COLUMN ==========")
print(df.dtypes)

# ============================================
# CHECK NULL VALUES
# ============================================

print("\n========== NULL VALUES IN EACH COLUMN ==========")
print(df.isnull().sum())

# ============================================
# REMOVE ROWS WITH NULL VALUES
# ============================================

df_rows_cleaned = df.dropna()

print("\n========== DATASET AFTER REMOVING NULL ROWS ==========")
print(df_rows_cleaned)

# ============================================
# REMOVE COLUMNS WITH NULL VALUES
# ============================================

df_columns_cleaned = df.dropna(axis=1)

print("\n========== DATASET AFTER REMOVING NULL COLUMNS ==========")
print(df_columns_cleaned)

# ============================================
# CONVERT DATA TYPES
# ============================================

# Keep Country Name as string
# Convert year columns to numeric

for column in df.columns[1:]:
    df[column] = pd.to_numeric(df[column], errors='coerce')

print("\n========== DATATYPES AFTER CONVERSION ==========")
print(df.dtypes)

# ============================================
# STATISTICAL ANALYSIS
# ============================================

print("\n========== STATISTICAL ANALYSIS ==========")
print(df.describe())

# ============================================
# MEDIAN POPULATION OF YEAR 2022
# ============================================

median_population = df['2022'].median()

print("\n========== MEDIAN POPULATION IN 2022 ==========")
print(median_population)

# ============================================
# MINIMUM POPULATION COUNTRY IN 2022
# ============================================

min_country = df.loc[df['2022'].idxmin(), 'Country Name']
min_population = df['2022'].min()

print("\n========== COUNTRY WITH MINIMUM POPULATION IN 2022 ==========")
print("Country Name :", min_country)
print("Population   :", min_population)

# ============================================
# MAXIMUM POPULATION COUNTRY IN 2022
# ============================================

max_country_2022 = df.loc[df['2022'].idxmax(), 'Country Name']
max_population_2022 = df['2022'].max()

print("\n========== COUNTRY WITH MAXIMUM POPULATION IN 2022 ==========")
print("Country Name :", max_country_2022)
print("Population   :", max_population_2022)

# ============================================
# POPULATION GROWTH CALCULATION
# ============================================

# Population Growth = Population in 2022 - Population in 1960

df['Population Growth'] = df['2022'] - df['1960']

print("\n========== POPULATION GROWTH ==========")
print(df[['Country Name', 'Population Growth']].head())

# ============================================
# TOP 10 COUNTRIES BY POPULATION IN 2022
# ============================================

top10 = df[['Country Name', '2022']].sort_values(by='2022', ascending=False).head(10)

print("\n========== TOP 10 COUNTRIES BY POPULATION IN 2022 ==========")
print(top10)

# ============================================
# BOTTOM 10 COUNTRIES BY POPULATION IN 2022
# ============================================

bottom10 = df[['Country Name', '2022']].sort_values(by='2022', ascending=True).head(10)

print("\n========== BOTTOM 10 COUNTRIES BY POPULATION IN 2022 ==========")
print(bottom10)

# ============================================
# FIND SPECIFIC COUNTRY DATA
# ============================================

country_name = "India"
country_data = df[df['Country Name'] == country_name]

print(f"\n========== DATA OF {country_name.upper()} ==========")
print(country_data)

# ============================================
# SAVE CLEANED DATASET
# ============================================

cleaned_file = "cleaned_population.csv"
df.to_csv(cleaned_file, index=False)

print("\n========== CLEANED DATASET SAVED SUCCESSFULLY ==========")
print("Saved File Name :", cleaned_file)

# ============================================
# PROGRAM COMPLETED
# ============================================

print("\n========== ANALYSIS COMPLETED SUCCESSFULLY ==========")



```
