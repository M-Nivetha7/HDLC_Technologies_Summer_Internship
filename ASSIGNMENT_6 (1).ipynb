```
from google.colab import files
import pandas as pd

# Upload CSV file
uploaded = files.upload()

# Get uploaded file name automatically
file_name = list(uploaded.keys())[0]

# Read CSV file into DataFrame
df = pd.read_csv(file_name)

# ==================================================
# TASK 1:
# Display first 5 rows and summary statistics
# ==================================================

print("First 5 Rows:")
print(df.head())

print("\nSummary Statistics:")
print(df.describe())

# ==================================================
# TASK 2:
# Select Customer_Age column and filter rows where
# Salary > 50000
# ==================================================

age_column = df['Customer_Age']

print("\nFirst 5 Values of Customer_Age:")
print(age_column.head())

filtered_df = df[df['Salary'] > 50000]

print("\nFirst 5 Rows where Salary > 50000:")
print(filtered_df.head())

# ==================================================
# TASK 3:
# Handle missing values
# Fill missing values in Customer_Age with mean age
# Drop rows with missing values in other columns
# ==================================================

mean_age = df['Customer_Age'].mean()

df['Customer_Age'] = df['Customer_Age'].fillna(mean_age)

other_columns = [col for col in df.columns if col != 'Customer_Age']

df = df.dropna(subset=other_columns)

print("\nMissing Values Handled Successfully")

# ==================================================
# TASK 4:
# Add total_price column and rename it to total_cost
# ==================================================

df['total_price'] = df['Unit_Price'] * df['Order_Quantity']

df.rename(columns={'total_price': 'total_cost'}, inplace=True)

print("\nDataFrame After Adding total_cost:")
print(df.head())

# ==================================================
# Display final DataFrame information
# ==================================================

print("\nUpdated Columns:")
print(df.columns.tolist())

print("\nFirst 5 Rows of Final DataFrame:")
print(df.head())
```
