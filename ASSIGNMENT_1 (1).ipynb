```python
# =====================================================
# DATA SCIENCE AND ML INTERNSHIP
# WEEK 1 ASSIGNMENT
# FULL UPDATED GOOGLE COLAB CODE
# =====================================================

# Import required libraries
import pandas as pd
import numpy as np

# =====================================================
# QUESTION 1
# Create DataFrame and count rows & columns
# =====================================================

# Creating the dataframe
data = {
    "Column1": [10, 20, 30],
    "Column2": [40, 50, 60],
    "Column3": [70, 80, 90],
    "Column4": [100, 110, 120]
}

# Create DataFrame
df1 = pd.DataFrame(data)

# Print DataFrame
print("DATAFRAME:\n")
print(df1)

# Count rows and columns
print("\nNumber of Rows:", df1.shape[0])
print("Number of Columns:", df1.shape[1])


# =====================================================
# QUESTION 2
# Read data.csv and perform operations
# =====================================================

# Upload CSV file in Google Colab
from google.colab import files

uploaded = files.upload()

# Read CSV file
df = pd.read_csv("data.csv")

# -----------------------------------------------------
# a. Print the dataset
# -----------------------------------------------------

print("\nFULL DATASET:\n")
print(df)

# -----------------------------------------------------
# b. Print first 5 rows
# -----------------------------------------------------

print("\nFIRST 5 ROWS:\n")
print(df.head())

# -----------------------------------------------------
# Data Cleaning
# -----------------------------------------------------

# Convert Area column to numeric
# Invalid values like 'XYZ' become NaN

df["Area"] = pd.to_numeric(df["Area"], errors="coerce")

# -----------------------------------------------------
# c. Remove NULL values
# -----------------------------------------------------

# Remove NULL values
df_cleaned = df.dropna()

print("\nDATA AFTER REMOVING NULL VALUES:\n")
print(df_cleaned)

# -----------------------------------------------------
# d. Statistical Analysis
# -----------------------------------------------------

print("\nSTATISTICAL ANALYSIS:\n")
print(df_cleaned.describe())

# -----------------------------------------------------
# Extra Analysis
# -----------------------------------------------------

# Check NULL values count
print("\nNULL VALUES IN EACH COLUMN:\n")
print(df.isnull().sum())

# Dataset information
print("\nDATASET INFORMATION:\n")
df.info()
```
